1. Importação dos dados e geração do relatório com Pandas Profiling:

```python
# Importando a biblioteca atualizada
import ydata_profiling as pp
import pandas as pd
import hvplot.pandas  # para gráficos interativos

# Carregar os dados e gerar o relatório
df = pd.read_excel('bitcoin_ historical_data3.xlsx')
profile = pp.ProfileReport(df, title="Relatório de Análise")
profile.to_file("relatorio1_pandas_profiling.html")
```

2. Geração de gráficos interativos usando hvPlot:

```python
df['High'] = pd.to_numeric(df['High'], errors='coerce')  # Converte para números, define 'NaN' em valores inválidos
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')  # Faz o mesmo para a coluna 'Price'

# Remove quaisquer linhas com valores inválidos (NaN) em 'High' ou 'Price'
df = df.dropna(subset=['High', 'Price'])

# Crie o gráfico interativo
grafico_interativo = df.hvplot.scatter(x='High', y='Price', title="Gráfico Interativo")

# Exiba o gráfico
grafico_interativo
```

3. Testando com o dataset do Pandas Profiling e explorando outros gráficos:

```python
# Importando as bibliotecas necessárias
import pandas as pd
import ydata_profiling as pp

# Testando com dados didáticos
df_vagas = pd.read_csv('netflix_titles.csv')

# Relatório Pandas Profiling
profile_vagas = pp.ProfileReport(df_vagas, title="Relatório de Análise")
profile_vagas.to_file("relatorio2_pandas_profiling.html")

# Explorando novos gráficos interativos
df_vagas.hvplot.hist('release_year', title="N° de filme por ano de lançamento")
```
